```python
df = pd.read_csv('analysis.csv')
df.head()
```

```python
bidder_counts = df['bidder'].value_counts().reset_index()
bidder_counts.columns = ['bidder', 'total_bids']
bidder_counts
```

```python
bidder_data = df.groupby('bidder', as_index=False)['win_count'].sum()
bidder_data
```

```python
win_rate_df = bidder_counts.merge(bidder_data, on='bidder')
win_rate_df
```

```python
win_rate_df['win_rate'] = win_rate_df['win_count'] / win_rate_df['total_bids']
win_rate_df.sort_values(by='win_rate', ascending=False, inplace=True)
win_rate_df
```

```python
print('Bidder '+ win_rate_df['bidder'].iloc[0] + ' has the highest Win Rate: ' + str(win_rate_df['win_rate'].iloc[0]))
```
